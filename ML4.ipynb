```python
current_x = 2 # starting point
previous_step_size = 1
precision = 0.000001 # tells us when to stop the algorithm
max_iter = 10000
iters = 0
lr = 0.01 #learning rate
```

```python
# y = (x+3)**2 and dy /dx = 2*(x+3)

def slope(x):
    return 2*(x+3)

def value_y(x):
    return (x+3)**2

x = []
y = []
x.append(current_x)
y.append(value_y(current_x))
```

```python
while precision < previous_step_size and iters < max_iter: #convergence criteria
    previous_x = current_x
    current_x = previous_x - lr * slope(previous_x)
    x.append(current_x)
    y.append(value_y(current_x))
    previous_step_size = abs(previous_x - current_x)
    print("Iteration: ", iters+1, "Value: ", current_x)
    iters += 1
```

```python
print("Local Minima for function is: ", current_x)
```
